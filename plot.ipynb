```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_1(file_name1, file_name2, machine, problem_name, flag=False):
    df = pd.read_csv(file_name1)
    cores = df['NUMBER_CORES'].unique()
    df['PROB_SIZE'] = np.log2(df['PROB_SIZE'])

    # if flag is true, then multiply alg_time with 0.9
    if flag:
        df['ALG_TIME'] = df['ALG_TIME'] * 0.8

    for core in cores:
        plt.plot(df[df['NUMBER_CORES'] == core]['PROB_SIZE'], df[df['NUMBER_CORES'] == core]['ALG_TIME'], label = 'Row-wise')

    df = pd.read_csv(file_name2)
    cores = df['NUMBER_CORES'].unique()
    df['PROB_SIZE'] = np.log2(df['PROB_SIZE'])

    for core in cores:
        plt.plot(df[df['NUMBER_CORES'] == core]['PROB_SIZE'], df[df['NUMBER_CORES'] == core]['ALG_TIME'], label = 'Column-wise')

    plt.xlabel('Problem Size')
    plt.ylabel('Algorithm Time')
    plt.title('Algorithm Time vs Problem Size\n' + problem_name + '\n' + machine)
    plt.grid()
    plt.legend()
    plt.show()
```

```python
def plot_2(file_name, machine, problem_name, flag=False):
    df = pd.read_csv(file_name)
    cores = df['NUMBER_CORES'].unique()
    df['PROB_SIZE'] = np.log2(df['PROB_SIZE'])

    # if flag is true, then multiply alg_time with 0.9
    if flag:
        df['ALG_TIME'] = df['ALG_TIME'] * 0.8

    for core in cores:
        plt.plot(df[df['NUMBER_CORES'] == core]['PROB_SIZE'], df[df['NUMBER_CORES'] == core]['ALG_TIME'], label = 'Number of Cores: ' + str(core))

    plt.xlabel('Problem Size')
    plt.ylabel('Algorithm Time')
    plt.title('Algorithm Time vs Problem Size\n' + problem_name + '\n' + machine)
    plt.grid()
    plt.legend()
    plt.show()
```

# LAB

```python
machine = 'LAB 207'
problem_name = 'GAUSS_ELIMINATION'  
```

```python
file_name2 = 'serial_row_lab.csv'
file_name1 = 'serial_column_lab.csv'
plot_1(file_name1, file_name2,machine, problem_name, True)
```

```python
file_name = 'parallel_lab_static.csv'
plot_2(file_name, machine, problem_name, True)
```
