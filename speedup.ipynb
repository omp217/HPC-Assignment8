```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_1(file_name, machine):
    plt.rcParams["figure.figsize"] = (8, 6)
    df = pd.read_csv(file_name)
    # print(df)
    
    # Store all the unique value of 'PORB_SIZE'
    porb_size = df['PROB_SIZE'].unique()

    # Store all the unique value of NUMBER_CORES
    number_cores = df['NUMBER_CORES'].unique()

    # If 0 is in number_cores, remove it
    if 0 in number_cores:
        number_cores = np.delete(number_cores, 0)

    # Iterate over the unique value of 'NUMBER_CORES'
    for i in number_cores:
        ans = []
        # Iterate over the unique value of 'PORB_SIZE'
        for j in porb_size:
           # Take the value of 'ALG_TIME' where 'NUMBER_CORES' is equal to i and 'PORB_SIZE' is equal to j
            temp = df.loc[(df['NUMBER_CORES'] == i) & (df['PROB_SIZE'] == j), 'ALG_TIME']
            # Take the value of 'ALG_TIME' where 'NUMBER_CORES' is equal to 0 and 'PORB_SIZE' is equal to j
            temp1 = df.loc[(df['NUMBER_CORES'] == 0) & (df['PROB_SIZE'] == j), 'ALG_TIME']
            # Calculate the speedup
            ans.append(temp1.values[0]/temp.values[0])
        # Take the log2 of prob_Size and store it in log_prob_size
        log_prob_size = np.log2(porb_size)
        # Plot the graph
        plt.plot(log_prob_size, ans, label = 'Number of cores = ' + str(i))

    # Set the title of the graph
    plt.title('Speedup for ' + machine)
    # Set the x-axis label
    plt.xlabel('PROB_SIZE')
    # Set the y-axis label
    plt.ylabel('Speedup')
    # Set the legend
    plt.legend()
    plt.grid()
    # Show the graph
    plt.show()
```
